# Introducing Threading
Now that we know that we have to use these 999 files for our task. We will be checking if using threading our performance do increases or not.<br>
**NOTE:** Becuase we can see that this task is I/O bound only , we are applying threading. If it was cpu bound then we could have used multiple processes.

```python
# importiting library needed for processing 
import concurrent.futures as conf
from time import time
# define a function to read the file 
def readfile(file):
    with open(file) as f:
        data = f.read()
    #file = file.replace('wiki/','').replace('.html','')
    return data
def checkperform(i):
    pool = conf.ThreadPoolExecutor(max_workers=i)
    #articles = []
    contents = pool.map(readfile,filenames)
    contents = list(contents)
# we will be checking the perfromance for 1, 2, 3, 4
times = []
for i in range(1,5):
    s = time()
    checkperform(i)
    times.append(time()-s)
    print("Time taken for worker={0} is {1}".format(i,times[i-1]))
```

This is for just one loop lets experiment with 1000 loops then take the mean of those values then see which no. of worker gives best response.

```python
times = {}
times[1] = []
times[2] = []
times[3] = []
times[4] = []
for j in range(1000):
    for i in range(1,5):
        s=time()
        checkperform(i)
        times[i].append(time()-s)
for k in times:
    m = sum(times[k])
    m /= len(times[k])
    print('Time taken for worker={} is {}'.format(k,m))
```

In above cell we can see that time taken for all the processes is nearly same, but minimum for one thread.

```python
# now lets read all files data in a list
pool = conf.ThreadPoolExecutor(max_workers=2)
contents = list(pool.map(readfile,filenames))
len(contents)
```

Now we will be parsing whole pages , as this task needs more cpu. So it is CPU bounded process. Hence we will be using multiprocesses rather than just using threads.

```python
contents[0]
```

```python
from bs4 import BeautifulSoup

def parser(html):
    soup = BeautifulSoup(html,'html.parser')
    return (str(soup.find_all('div',id='content')[0]))
start = time()
pool = conf.ProcessPoolExecutor(max_workers=2)
parsed = pool.map(parser,contents)
parsed = list(parsed)
print(time()-start)
```

We can see this process took around 27.94 seconds for 2 workers or parallel processing using 2 processors. Let's try to make a function and compare timing.

```python
def time_proc(i):
    s = time()
    pool = conf.ProcessPoolExecutor(max_workers=i)
    pool.map(parser,contents)
    print("Proces with {} worker took {} seconds".format(i,time()-s))
for i in range(1,4):
    time_proc(i)
```

# Finding the most common tags
Since in this process also we have to search in every text we will have to use multiple processes as it is cpu expensive task.

```python
import concurrent.futures as conf
import time
def common_tag(html):
    soup = BeautifulSoup(html,'html.parser')
    tag_dict = {}
    for tag in soup.find_all():
        tag = tag.name
        if tag in tag_dict:
            tag_dict[tag] += 1
        else:
            tag_dict[tag] = 1
    return tag_dict
start = time.time()
pool = conf.ProcessPoolExecutor(max_workers=3)
dicts = pool.map(common_tag,parsed)
dicts = list(dicts)
tag_dict = {}
for dic in dicts:
    for d in dic:
        if d in tag_dict:
            tag_dict[d] += dic[d]
        else:
            tag_dict[d] = dic[d]
total = time.time() - start
print(total)
tag_dict
```

# Counting the common words

```python
from collections import Counter
import re
def common_words(html):
    soup = BeautifulSoup(html,'html.parser')
    text = soup.get_text()
    #text = text.lower()
    text = re.sub('\W+','',text.lower())
    words = [w for w in text.split(' ') if len(w)>5]
    return Counter(words).most_common(10)
start = time.time()
pool = conf.ProcessPoolExecutor(max_workers=3)
counts = list(pool.map(common_words,parsed))
counter = counts[0]
for c in range(1,len(counts)):
    counter = counter + counts[c]
common_words_ten = counter.most_common(10)
total = time.time() - start
print(total)
common_words_ten
```
