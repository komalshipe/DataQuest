# Visualization
Now that we have a good dataset of 172 rows without any null values now we can focus on visualization. In following section we are going to visualize the scatter plots of following relations
- `Sample_size vs Median`
- `Sample_size vs Unemployment_rate`
- `Full_time vs Median`
- `ShareWomen vs Unemployment_rate`
- `Men vs Median`
- `Women vs Median`

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter')
plt.show()
```

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter')
plt.show()
```

In above plot we can see that Unemployment_rate is more clustered between sample size 0 to 500.

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter')
plt.show()
```

In above plot we can see people who are employed for full_time most of them have salary less than 60,000. Only one entry is showing salary over 100000.

```python
ax = recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
plt.show()
```

In above plot we can observe unemployment rate and women share is nearly constant.(ShareWomen is defined for majors so irrespective of major unemployment rate is constant for women)

```python
recent_grads.plot(x='Men',y='Median',kind='scatter')
plt.show()
```

```python
recent_grads.plot(x='Women',y='Median',kind='scatter')
plt.show()
```

## Let's check whetere students from popular majors  make more money or not?
We will take following steps
1. First sort the dataset according to Popularity(Total)
2. Plot the Major against median salary<br>
**Note - Since Major is categorical we will show against the index values of new df**

```python
# first lets make a new df with values sorted by Total column in ascending order
pop_df = recent_grads.sort_values('Total',ascending=True)
pop_df.head()
```

Ohhh! The index column has values which are not order so it will be not a good to put it like that way so lets fix it.

```python
pop_df = pop_df.reset_index(drop=True)
pop_df[['Major','Total']].head()
```

```python
pop_df[['Major','Total']].tail()
```

Now this looks cool. Let's plot histogram and line graph.

```python
ax = pop_df.reset_index().plot(x='index',y='Median',kind='bar')
#ax.set_xlabel([i for i in range(pop_df.shape[0])])
ax.set_xticks([0,pop_df.shape[0]])
plt.show()
```

```python
pop_df.reset_index().plot(x='index',y='Median',kind='line')
plt.show()
```

So by observing graph we can roughly esitmate that the less popular majors have more salary compared to majors which have more students.

## Let's check whether females make more money where the share of women was more in those majors are highly salaried or not
1. First sort the df with respect to sharewomen
2. Plot the salary median and sharewomen in line graph

```python
# we will do all steps in one line only as we have done it earlier
women_df = recent_grads.sort_values('ShareWomen').reset_index(drop=True)
women_df[['Median','ShareWomen']].head()
```

```python
women_df.plot(x='ShareWomen',y='Median')
plt.show()
```

The graph is indicating that the majors which has women share less are highly paid and as women salary increases the salary decreases. **This indicates that the majors in which women share is high are less paid.**

### Comparison of jobs for male and female
Now we will compare jobs for female first then male candidates

```python
recent_grads[:10].plot.bar(x='Major',y='ShareWomen')
recent_grads.tail(10).plot.bar(x='Major',y='ShareWomen')
plt.show()
```

Here we can see the majors in which women share is high are low paying.

```python
recent_grads[:10].plot.bar(x='Major',y='Unemployment_rate')
recent_grads[recent_grads.shape[0]-10:].plot.bar(x='Major',y='Unemployment_rate')
plt.show()
```
